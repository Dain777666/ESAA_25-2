## 1.1 인공지능, 머신러닝과 딥러닝

< 머신러닝 VS 딥러닝 >

- 머신러닝: 인간이 특성 추출을 담담, 컴퓨터가 분류 작업을 처리

- 딥러닝: 컴퓨터가 특성 추출과 분류 작업을 모두 처리

## 1.2 머신러닝이란

### 1.2.1 머신러닝 학습 과정

- 머신러닝은 크게 학습과 예측 단계로 나누어짐

- 특성 추출: 컴퓨터가 데이터를 분석하여 일정한 규칙을 찾아내려면, 사람이 인지하는 데이터를 컴퓨터가 인지할 수 있는 데이터로 변환해야 함. 이때, 데이터별로 어떤 특징을 가지고 있는 지 찾아내고 이를 토대로 데이터를 벡터로 변환하는 작업.

image.png



## 1.3 딥러닝이란

- 딥러닝과 머신러닝의 큰 차이점은 딥러닝은 인간의 뇌를 기초로 하여 설계했다는 점.

### 1.3.1 딥러닝 학습 과정

1. 데이터 준비: 파이토치에서 제공하는 데이터셋이나 캐글의 공개 데이터 이용

2. 모델 정의: 신경망 생성 단계. 일반적으로 은닉층 개수가 많을수록 성능이 좋아지지만, 과적합이 발생할 확률이 높음.

3. 모델 컴파일: 활성화 함수, 손실 함수, 옵티마이저를 선택함. 옵티마이저는 손실 함수를 기반으로 네트워크 업데이트 방법을 결정함.

4. 모델 훈련: 한 번에 처리할 데이터양을 지정함. 배치와 훈련의 횟수인 에포크 선택이 중요함.

- '훈련 데이터셋 1000개에 대해 배치 크기와 에포크는 각각 20, 10' -> 1000/20*10 = 총 500번 가중치가 업데이트 됨.

5. 모델 예측: 검증 데이터셋을 생성한 모델에 적용하여 실제 예측 진행

### 1.3.2 딥러닝 학습 알고리즘

- 지도 학습, 비지도 학습, 전이 학습, 강화 학습으로 분류됨

< 지도 학습 >

- 합성곱 신경망(CNN): 이미지 분류, 이미지 인식, 이미지 분할로 분류됨

- 순환 신경망(RNN): 시계열 데이터를 분류할 때 사용됨. 역전파 과정에서 기울기 소멸 문제가 발생하는 단점이 있음. -> 해결방안: LSTM

- LSTM(Long Short-Term Memory): 망각 게이트(과거 정보를 잊기 위한 게이트), 입력 게이트(현재 정보를 기억하기 위한 게이트), 출력 게이트를 도입하여 기울기 소멸 문제 해결함.

< 비지도 학습 >

- 워드 임베딩과 군집으로 분류됨

- 워드 임베딩: 단어를 벡터로 표현하는 것. 자연어 처리 분야의 일종으로 번역이나 음성 인식 등 서비스에서 사용함.

- 군집: 한 클러스터 안의 데이터는 매우 비슷하게 구성하고 다른 클러스터의 데이터와 구분되도록 나눔

< 전이 학습>

- 사전에 학습이 완료된 모델을 가지고 우리가 원하는 학습에 미세 조정 기법을 이용하여 학습시키는 방법

- 사전 학습 모델: 풀고자 하는 문제와 비슷하면서 많은 데이터로 이미 학습이 되어 있는 모델

## 2.1 파이토치 개요

**파이토치**: GPU에서 텐서 조작 및 동적 신경망 구축이 가능한 프레임워크

- GPU: 병렬연산에서 효과적임

- 텐서: 파이토치의 데이터 형태, 단일 데이터 형식으로 된 자료들의 다차원 행렬

- 동적 신경망: 훈련을 반복할 때마다 네트워크 변경이 가능한 신경망. 예를 들어, 학습 중에 은닉층을 추가하거나 제거하는 등 모델의 네크워크 조작이 가능함

- torch.tensor로 구현함

````python
import torch.nn as nn

def MLP(in_features=1, hidden_features=20, out_features=1):
  hidden = nn.Linear(in_features=in_features, out_features=hidden_features)
  activation = nn.Relu()
  output = nn.Linear(in_features=hidden_features, out_features=out_features,
                     bias=True)
  net = nn.Sequential(hidden, activation, output)
  return net
````
