## 9.2 가우시안 혼합

- 가우시안 혼합 모델(GMM): 샘플이 파라미터가 알려지지 않은 여러 개의 혼합된 가우시안 분포에서 생성되었다고 가정하는 확률 모델

- 하나의 가우시안 분포에서 생성된 모든 샘플은 하나의 클러스터를 형성

- 생성모델이므로 모델에서 새로운 샘플을 만들 수 있음

< GaussianMixture 클래스 >

- 기댓값-최적화 알고리즘을 사용함(k-평균과 공통점 많음)

- 클러스터 파라미터를 랜덤하게 초기화하고 수렴할 때까지 두 단계를 반복

- 기댓값 단계: 샘플을 클러스터에 할당

- 최대화 단계: 클러스터를 업데이트(책임이 가장 많은 샘플에 크게 영향을 받음)

- 소프트 클러스터 할당을 사용함(확률 기반)

- 책임: 클러스터에 속할 추정 확률

- 최적의 솔루션으로 수렴하기 어려운 경우 -> 알고리즘의 학습 파라미터 개수 제한(클러스터의 모양과 방향의 범위를 제한)



```python
import numpy as np
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000,noise=0.05)
```

```python
from sklearn.mixture import GaussianMixture

gm = GaussianMixture(n_components=3, n_init=10)
gm.fit(X)
```

```python
gm.weights_
```

```python
gm.means_
```

```python
gm.covariances_
```

```python
gm.converged_
```

```python
gm.n_iter_
```

```python
gm.predict(X) # 하드 군집
```

```python
gm.predict_proba(X) # 소프트 군집
```

```python
X_new, y_new = gm.sample(6)
X_new
```

```python
y_new
```

```python
# 모델의 밀도 추정
# 점수가 높을수록 밀도가 높음
gm.score_samples(X)
```

## 9.2.1 가우시안 혼합을 사용한 이상치 탐지

- 밀도가 낮은 지역에 있는 샘플 -> 이상치

- 사용할 밀도 임곗값을 정해야 함

- 특이치 탐지: 이상치로 오염되지 않은 깨끗한 데이터셋에서 훈련함

```python
densities = gm.score_samples(X)
density_threshold = np.percentile(densities, 4)
anomalies = X[densities < density_threshold]
```

## 9.2.2 클러스터 개수 선택하기

- BIC(Bayesian Information Criterion) 혹은 AIC(Akaike Information Criterion) 최소화

  image.png

- 학습할 파라미터가 많은 모델에게 벌칙을 가하고 데이터에 잘 학습하는 모델에게 보상을 더함

< 가능도 함수 >

- 확률: 파라미터 theta인 확률 모델이 주어지면, 미래 출력 x가 얼마나 그럴듯한지 설명함(x의 함수) -> 확률분포 적분하면 1

- 가능도: 출력 x를 알고 있을 때, 특정 파라미터 값 theta가 얼마나 그럴듯한지 설명함(theta의 함수) -> 확률분포 적분하면 양숫값

- 일반적으로 로그 가능도를 최대화함

```python
gm.bic(X)
```

```python
gm.aic(X)
```

## 9.2.3 베이즈 가우시안 혼합 모델

< BayesGaussianMixture 클래스 >

- 불필요한 클러스터의 가중치를 0으로 만듦

- 클러스터 개수 'n_components'를 최적의 클러스터 개수보다 크다고 믿을만한 값으로 지정

- 클러스터 파라미터는 고정된 모델 파라미터가 아니라 클러스터 할당처럼 잠재 확률 변수

  image.png

< 베타 분포 >

- 고정 범위 안에 놓인 값을 가진 확률 변수를 모델링할 때 자주 사용됨

- SBP: 새로운 샘플이 작은 클러스터보다 큰 클러스터에 합류할 가능성이 높은 데이터셋이 잘 맞는 모델
